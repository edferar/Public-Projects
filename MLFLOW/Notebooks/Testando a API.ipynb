```python
import requests
```

```python
url = "http://127.0.0.1:8081/invocations"
```

```python
import pandas as pd

df = pd.read_csv("../data/Housing.csv")
data = df[["bedrooms", "bathrooms", "sqft_living",
       "sqft_lot", "floors", "waterfront", "view", "condition", "grade",
       "sqft_above", "sqft_basement", "yr_built", "yr_renovated", 
       "lat", "long", "sqft_living15", "sqft_lot15"]].copy()
```

```python
header = {"Content-Type":"application/json"}
data_request = {"dataframe_split": data[:1].to_dict(orient="split", index=False)}
response = requests.post(url,json=data_request,headers=header)
print(response.status_code, response.text)
```

```python
header = {"Content-Type":"application/json"}
data_request = {"dataframe_split": data.to_dict(orient="split", index=False)}
response = requests.post(url,json=data_request,headers=header)
print(response.status_code, response.text)
```
